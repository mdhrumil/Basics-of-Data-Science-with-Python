# Data Visualization - Python

## This notebook documents some insightful visualizations using seaborn library in Python. Some of the visualizations includes histograms, kernel density plots, box plots, violin plots, heatmaps and regression plots.

### Installing Seaborn library

Use command "pip install seaborn" to install the seaborn library or you can also use command "conda install seaborn " if you have Anaconda Prompt installed. 

```python
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
from numpy.random import randn
%matplotlib inline
```

### Histograms

```python
#Generating 100 random values
sample = randn(100)
plt.figure(figsize=(10,10))
plt.hist(sample)
```

```python
sample2 = randn(50)
plt.figure(figsize=(10,10))
plt.hist(sample,color="indianred")
```

Two histograms in one plot

```python
plt.figure(figsize=(10,10))
plt.hist(sample,density=True,color="blue",alpha=0.5,bins=20)
plt.hist(sample2,density=True,color="indianred",alpha=0.5,bins=20)
```

 The density parameter in the hist function sets the plots to be normalized in a probability distribution. 

### Jointplots

```python
data1 = randn(100)
data2 = randn(100)
plt.figure(figsize=(15,15))
sns.jointplot(data1,data2)
```

Hex bins for representing the scatter

```python
plt.figure(figsize=(15,15))
sns.jointplot(data1,data2,kind="hex",height=7,color="red")
```

### Kernel density estimation plots

A rugplot

```python
dataset = randn(50)
plt.figure(figsize=(10,10))
sns.rugplot(dataset)
```

```python
plt.figure(figsize=(10,10))
sns.kdeplot(dataset)
```

```python
dataset1 = randn(50)
dataset1 = dataset1 - dataset
plt.figure(figsize=(10,10))
sns.kdeplot(dataset, shade=True,color="yellow",kernel="gau")
sns.kdeplot(dataset1,shade=True,color="indianred",kernel="cos")
```

KDE with Jointplots

```python
plt.figure(figsize=(15,15))
sns.jointplot(data1,data2,kind="kde",shade=True,color="violet")
```

### Combining plots styles using distplot

By default, the distplot combines KDE over corresponding histogram.

```python
plt.figure(figsize=(10,10))
dataset1 = randn(100)
sns.distplot(dataset1,bins=25)
```

Switching histogram with a rugplot

```python
plt.figure(figsize=(10,10))
sns.distplot(dataset1,rug=True,hist=False,bins=25)
```

Passing specific plot controls:

```python
plt.figure(figsize=(10,10))
sns.distplot(dataset1,bins=20,kde_kws={'color':'indianred','label':'KDE Plot'},
             hist_kws={'color':'blue','label':'Histogram','alpha':0.3})
```

### Box plots and Violin plots

```python
data1 = randn(100)
data2 = randn(100)+9
plt.figure(figsize=(10,10))
sns.boxplot(data1,orient="v",width=0.4,color="indianred")
sns.boxplot(data2,orient="v",width=0.4,color="yellow")
```

Corresponding violinplots

```python
plt.figure(figsize=(10,10))
sns.violinplot(data1,color="indianred",orient="v",width=0.4)
sns.violinplot(data2,color="yellow",orient="v",width=0.4)
```

### Regression Plots

Seaborn provides some sample datasets in form of pandas dataframes, so we'll use them for further plots.

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize=(15,15))
sns.lmplot("total_bill","tip",tips)
```

Controlling individual plots for their properties

```python
plt.figure(figsize=(15,15))
sns.lmplot("total_bill","tip",tips, scatter_kws={"marker": "o", "color": "indianred"},line_kws={"linewidth": 1, "color": "blue"})
```

Plotting higher order polynomial trends

```python
plt.figure(figsize=(15,15))
sns.lmplot("total_bill", "tip", tips,order=3,scatter_kws={"marker": "o", "color": "indianred"},line_kws={"linewidth": 1, "color": "blue"})
```

```python
plt.figure(figsize=(15,15))

sns.lmplot("total_bill", "tip", tips,order=5,scatter_kws={"marker": "o", "color": "indianred"},line_kws={"linewidth": 1, "color": "blue"})
```

```python
# lmplot() also works on discrete variables, such as the percentage of the tip
# Create a new column for tip percentage
plt.figure(figsize=(15,15))
tips["tip_pect"]=100*(tips['tip']/tips['total_bill'])
sns.lmplot("total_bill", "tip_pect", tips, hue="day")
```

### Heatmaps and Clustered Matrices

```python
flight = sns.load_dataset("flights")
```

```python
flight.head()
```

```python
#Pivoting the dataset 
flight = flight.pivot("month","year","passengers")
flight
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(flight,linewidth=0.1,annot=True)
```

```python
# We can choose our own 'center' for our colormap
plt.figure(figsize=(15,15))
sns.heatmap(flight,center=flight.loc['August',1954],linewidth=0.1,annot=True)
```

Cluster matrices

```python
plt.figure(figsize=(20,20))
sns.clustermap(flight)
```
